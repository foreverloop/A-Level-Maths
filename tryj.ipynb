```python
#symbolic differentiation
from sympy import * #dangerous to use *, don't do this unless there is a good reason
import numpy as np
x = Symbol('x')
y = x**5 + 2
yprime = y.diff(x)
print yprime

#lambdify the function so it can be used by python
f = lambdify(x, yprime, 'numpy')
print f(2)
print 5*2**4
```

```python
import sympy as sym
import numpy as np
from matplotlib import pyplot as plt

x = sym.Symbol('x')
y = x**3 + 3
yprime = y.diff(x)
print "the derivative function is: {0}".format(yprime)
funyprime = sym.lambdify(x,yprime,'numpy')

x_axis = range(1,31)
y_axis_values = [(x**3 + 3) for x in range(1,31)]
y_axis_derivative = [funyprime(x) for x in range(1,31)]
print sum(y_axis_derivative)
plt.plot(x_axis,y_axis_values,color='purple')
plt.plot(x_axis,y_axis_derivative,color='green')
plt.show()
```
